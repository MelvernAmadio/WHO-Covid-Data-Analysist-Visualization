```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("who_covid_19.csv")
df.head(10)
```

```python
df.rename(columns =  {'Name': 'Country'}, inplace = True)
```

```python
df['Country']= df['Country'].replace('United States of America', 'United States')
df['Country']= df['Country'].replace('Russian Federation', 'Russian')
df['Country']= df['Country'].replace('The United Kingdom' , 'United Kingdom')
df['Country']= df['Country'].replace('Iran (Islamic Republic of)' , 'Iran')
df['Country']= df['Country'].replace('Czechia' , 'Czech Republic')
df['Country']= df['Country'].replace('occupied Palestinian territory, including east Jerusalem' , 'Palestine, State of')
df['Country']= df['Country'].replace('Republic of Moldova' , 'Moldova')
df['Country']= df['Country'].replace('Viet Nam' , 'Vietnam')
df['Country']= df['Country'].replace('Bolivia (Plurinational State of)' , 'Bolivia')
df['Country']= df['Country'].replace( 'Palestine','Palestine, State of')
df['Country']= df['Country'].replace('Venezuela (Bolivarian Republic of)' , 'Venezuela')
df['Country']= df['Country'].replace('Republic of Korea' , 'Korea, Republic of')
df['Country']= df['Country'].replace( 'Korea(Republic of','Korea, Republic of')
df['Country']= df['Country'].replace( 'Bosnia and Herzegovina','Bosnia And Herzegovina')
df['Country']= df['Country'].replace( ' Namibia ','Namibia')
df['Country']= df['Country'].replace( 'Syrian Arab Republic','Syria')
df['Country']= df['Country'].replace( 'United Republic of Tanzania','Tanzania')

```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#Handle Missing Value
for column in df.select_dtypes(include=np.number).columns:
    df[column] = df[column].fillna(df[column].mean())
df.isnull().sum()
```

```python
df = df[df['Country'] != 'Global']
df =df[df['Country'] != 'Other']
```

```python
df.isnull().sum()
```

# **Visualization Analysist**

### Kasus Covid-19 di Setiap Benua

```python
df = pd.DataFrame(df)
df_grouped = df.groupby('WHO Region')['Cases - cumulative total'].sum().reset_index()
df_grouped_sorted = df_grouped.sort_values(by='Cases - cumulative total', ascending=False)

df_grouped_sorted['Rank'] = df_grouped_sorted['Cases - cumulative total'].rank(ascending=False, method='min').astype(int)
print(df_grouped_sorted)
```

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']
df.groupby('WHO Region')['Cases - cumulative total'].sum().plot.bar(title="Total Cases per Region",ylabel='Cases - cumulative total (Dalam Puluhan Juta)', color=colors)
```

Dari hasil sorting dan Visualisasi yang dilakukan, kita dapat menyimpulkan bahwa untuk setiap region, kasus Covid-19 paling banyak terjadi di benua Amerika dengan total kasus kumulatif mencapai 83.479.049 kasus. Lalu untuk benua dengan kasus terendah adalah Benua Afrika, dimana total kasus Covid-19 kumulatif mencapai 5.620.667 kasus

### Kasus Kematian Akibat Covid-19 di Setiap Benua

```python
df = pd.DataFrame(df)
df_grouped = df.groupby('WHO Region')['Deaths - cumulative total'].sum().reset_index()
df_grouped_sorted = df_grouped.sort_values(by='Deaths - cumulative total', ascending=False)

df_grouped_sorted['Rank'] = df_grouped_sorted['Deaths - cumulative total'].rank(ascending=False, method='min').astype(int)
print(df_grouped_sorted)
```

```python
colors = ['#00008B', '#8B0000', '#006400', '#9400D3', '#FF8C00', '#008B8B']
df.groupby('WHO Region')['Deaths - cumulative total'].sum().plot.bar(title="Total Death per Region",ylabel='Deaths - cumulative total (Dalam Jutaan)', color = colors)
```

Dari hasil sorting dan Visualisasi yang dilakukan, kita dapat menyimpulkan bahwa untuk setiap region, kasus kematian akibat Covid-19 paling banyak terjadi di benua Amerika dengan total kasus kumulatif mencapai 2.097.836 kasus. Lalu untuk benua dengan kasus terendah adalah Benua Pasifik Barat, dimana total kasus Covid-19 kumulatif mencapai 88.975 kasus

### Europe

```python
new_df_europe = df[df["WHO Region"].isin(['Europe'])].sort_values('Deaths - cumulative total per 100000 population')
```

```python
euro = sns.barplot(x=new_df_europe['Country'], y='Deaths - cumulative total per 100000 population', data=new_df_europe, palette ='Set2')
euro.set_xticklabels(euro.get_xticklabels(), rotation=90)
euro.set_title('Europe')
```

### Americas

```python
new_df_americas = df[df["WHO Region"].isin(['Americas'])].sort_values('Deaths - cumulative total per 100000 population')
```

```python
amer = sns.barplot(x=new_df_americas['Country'], y='Deaths - cumulative total per 100000 population', data=new_df_americas,palette='viridis')
amer.set_xticklabels(amer.get_xticklabels(), rotation=90)
amer.set_title('Americas')
```

## Total Covid-19 Cases By Country

```python
import plotly.graph_objects as go
import plotly.offline as po
```

```python
CasecountryData = dict(
                    type = 'choropleth',
                    colorscale="Bluyl",
                    locations = df['Country'],
                    locationmode = 'country names',
                    z = df['Cases - cumulative total'],
                    text = 'Case(s)',
                    colorbar = {'title' : 'Case(s)'}
                    )
layout = dict(title = 'Total COVID 19 Cases by Country',
              geo = dict(showframe = False,
                       projection = {'type':'orthographic'},
                       showlakes = True, lakecolor="blue",
                       showocean=True, oceancolor="LightBlue",
                       showland=True, landcolor="LightGreen"))
choromap = go.Figure(data = [CasecountryData],layout = layout)
po.iplot(choromap,validate=False)
```

```python
DeathcountryData = dict(
                    type = 'choropleth',
                    colorscale="Agsunset",
                    locations = df['Country'],
                    locationmode = 'country names',
                    z = df['Deaths - cumulative total'],
                    text = 'Death(s)',
                    colorbar = {'title' : 'Death(s)'}
                    )
layout = dict(title = '<b>Total COVID 19 Deaths by Country</b>',
              geo = dict(showframe = False,
                       projection = {'type':'orthographic'},
                       showlakes = True, lakecolor="blue",
                       showocean=True, oceancolor="LightBlue",
                       showland=True, landcolor="LightGreen"))
choromap = go.Figure(data = [DeathcountryData],layout = layout)
po.iplot(choromap,validate=False)
```
